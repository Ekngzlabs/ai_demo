```python
class Student:
    def __init__(self, name, age):
        """Initialise a new Student with a name and age."""
        self._name = name   # create the "name" attribute on this object
        self._age  = age    # create the "age" attribute on this object
        
    @property
    def age(self):
        return self._age
        
    @age.setter
    def age(self,age):
        self._age = age
        
# Create objects — the arguments go to __init__'s parameters
s1 = Student("Amina", 20)
s2 = Student("Juma",  22)

s1.age = 35
# Access attributes using dot notation
print(s1._name, s1.age)   # Amina  20
print(s2._name, s2._age)   # Juma   22
```

```python
class BankAccount:
    def __init__(self,account_holder: str, balance = 0.0):
        self.account_holder = account_holder
        self._balance = balance

    def __str__(self):
        return f"BankAccount({self.account_holder}, balance: {self._balance})"
        
    def deposit(self,amount):
        if amount > 0:
            self._balance += amount
            return True
        else:
            print(f"Invalid Amount: {amount}")
            return False

    def withdraw(self,amount):
        if amount < 0:
            print(f"Invalid Amount: {amount}")
            return False
        elif amount > self._balance:
            print("Insuficient Balance!")
            return False
        else: 
            self._balance -= amount
            return True

    def get_balance(self):
        return self._balance
```

```python
acc1 = BankAccount("Juma Asukile")

assert acc1.deposit(10000) == True, "Deposit Failed"
assert acc1.deposit(-10000) == False, "Deposit Failed"
assert acc1.deposit(0) == False, "Deposit Failed"
assert acc1.withdraw(5000) == True, "Withdraw Failed"
assert acc1.withdraw(10000) == False, "Withdraw Failed"
assert acc1.withdraw(-10000) == False, "Withdraw Failed"
```

```python
class Sensor:
    MIN_TEMP, MAX_TEMP = -50.0, 100.0

    def __init__(self, temp=25.0):
        self._temperature = temp

    @property
    def temperature(self):
        """Getter — accessed as  sensor.temperature  (no parentheses)."""
        return self._temperature

    @temperature.setter
    def temperature(self, value):
        """Setter — triggered by  sensor.temperature = 30  ."""
        if Sensor.MIN_TEMP <= value <= Sensor.MAX_TEMP:
            self._temperature = value
        else:
            print(f"Error: {value} out of range.")
```

```python
class SensorArray:
    def __init__(self):
        self.sensor_list = []

    # add_sensor(sensor) appends a Sensor to the internal list
    def add_sensor(self, sensor):
        if isinstance(sensor, Sensor):
            self.sensor_list.append(sensor)
        else:
            print(f"{sensor} - is not a Sensor object")

    # average_temperature() returns the mean temperature across all sensors
    def average_temperature(self):
        if len(self.sensor_list) <= 0:
            return 0.0
        return sum([sensor.temperature for sensor in self.sensor_list]) / len(self.sensor_list)

    # alert() returns a list of sensor indices where temperature > 37.5 (elevated)
    def alert(self):
        alert_list = []
        for index,sensor in enumerate(self.sensor_list):
            if sensor.temperature > 37.5:
                alert_list.append(index)
        return alert_list
            
```

```python
sensor_array = SensorArray()
# Test with 5 sensors at temperatures [36.1, 37.8, 35.9, 38.2, 37.0]. Verify that alert() returns [1, 3].
for temp in [36.1, 37.8, 35.9, 38.2, 37.0]:
    sensor_array.add_sensor(Sensor(temp))

sensor_array.alert()
```
