### 7 7. Write a complete Python program that: (a) asks the user to enter 5 subject marks (0–100 each, validated with a loop); (b) calculates total, average, highest, and lowest marks; (c) uses a for loop to print each mark alongside a pass/fail judgment (pass >= 50); (d) validates your program against at least one boundary case and one edge case

```python
marks = []
```

```python
def validate(mark):
    if not 0 <= mark <= 100:
        print('Invalid! marks should be between 0 and 100')
        return False
    else: return True
```

```python
def get_mark():
    mark = float(input("Enter Mark: "))
    if validate(mark):
        return mark
    else: 
        return get_mark()
```

```python
for i in range(1,6):
    mark = get_mark()
    marks.append(mark)
```

```python
print(marks)
```

```python
total = sum(marks)
print(f"Total: {total}")
```

```python
average = total / len(marks)
print(f"Average: {average}")
```

```python
print(f"Highest: {max(marks)}, Lowest: {min(marks)}")
```

```python
def judge(mark):
    if (mark >= 50):
        print(f"{mark} - Pass")
    else:
        print(f"{mark} - Fail")
        
```

```python
for mark in marks:
    judge(mark)
```
